```python
train, test = pd.read_csv('Linearized_reduced_train.csv'), pd.read_csv('Linearized_reduced_test.csv')
```

```python
train.iloc[:,-1] = train.iloc[:,-1].apply(lambda x: 1-x)
```

```python
train_X = train.iloc[:,1:-1]
train_Y = train.iloc[:,-1]
test_X = test.iloc[:,1:]
```

```python
xgb_model = XGBClassifier(n_jobs = 7)

parameters = {'max_depth': [7],
              'n_estimators': [450, 600], 
              'reg_alpha':[0.5],
              'reg_lambda':[1],
              'seed': [42]}

clf = GridSearchCV(xgb_model, param_grid=parameters, scoring='precision', verbose=5)


clf.fit(train_X, train_Y)
```

```python
#train.iloc[:,-1] = train.iloc[:,-1].apply(lambda x: 1-x)
reg = clf.best_estimator_
reg.fit(train.iloc[:,1:-1], train.iloc[:,-1])
predictions = reg.predict(test_X)
model = reg
```

### Submission

```python
def adjust_value(confidence, default_ind, threshold):
    if default_ind == 0:
        return default_ind
    else:
        if confidence > threshold:
            return default_ind
        else:
            return (1-default_ind)
```

```python
def probability_dataframe(model,test_data, predictions, threshold):
    probabilities = model.predict_proba(test_data)
    d = {'application_key':test['application_key'], 'default_ind':predictions.tolist(),\
         'probability':probabilities.tolist()}

    prediction_df = pd.DataFrame(data = d)
    #print(prediction_df)
    prediction_df['Adjusted_Values'] = prediction_df.apply(lambda row: adjust_value(row[2][int(row[1])], int(row[1]),threshold), axis = 1)
    prediction_df['Confidence'] = prediction_df.apply(lambda row: row[2][row[3]], axis =1)
    prediction_df.sort_values(by = 'Confidence', axis = 0, ascending=False, inplace = True)
    prediction_df.reset_index(drop=True, inplace = True)
    
    return prediction_df
```

```python
probability_df = probability_dataframe(model, test_X, predictions, threshold=0.99)
probability_df.head()
```

```python
feature_importances = pd.DataFrame(model.feature_importances_,
                                   index = train_X.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```
